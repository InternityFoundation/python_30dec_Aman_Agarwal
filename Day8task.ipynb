```python
import pandas_datareader.data as web
all_data = {ticker: web.get_data_yahoo(ticker)
for ticker in ['AAPL', 'IBM', 'MSFT', 'GOOG']}
price = pd.DataFrame({ticker: data['Adj Close']
for ticker, data in all_data.items()})
volume = pd.DataFrame({ticker: data['Volume']
for ticker, data in all_data.items()})
```

```python
print(price)
```

```python
print(volume)
```

```python
returns=price.pct_change()
print(returns)
```

```python
returns.tail()
```

```python
returns.corr() #it is defining the correlation between each and every variable 
```

```python
#we can also find the correlation between any particular variable also
returns.AAPL.corr(returns.AAPL) #it is finding the correlation between the AAPl AND AAPL
```

```python
returns.IBM.corr(returns.MSFT) #in this we can use the corr method of the python dataframes
```

```python
returns.cov() #it is finding the covariance between each of the vaiable
```

```python
#we can also find the covariance between a particular variable also
returns.AAPL.cov(returns.AAPL)
```

```python
#we can find the pairwise correaltion by the corrwith method
returns.corrwith(returns.AAPL)
```

```python
returns.covwith(returns.AAPL) #datafarme does not have any covwith function
```

```python
#by passing axis='columns' we can do this thing by the column wise instead of rowwise
returns.corrwith(returns.IBM,axis='columns')
```

```python
###Unique values, values_counts and membership
objk=pd.Series(['a','a','a','b','b','c','c','d','d','d','d','d'])
objk.unique()
```
